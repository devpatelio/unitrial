```python
import requests

class ClinicalTrialDataset():
    def __init__ (self, key_words, CLINICAL_ENDPOINT):
        self.CLINICAL_ENDPOINT = CLINICAL_ENDPOINT
        self.trial_parameters = [
            'NCTId',
            'Title',
            'Recruitment Status',
            'Conditions',
            'Interventions',
            'Study Type',
            'Brief Summary',
            'Intervention Description',
            'Eligibility Criteria',
            'Primary Outcomes',
            'Secondary Outcomes',
            'Start Date',
            'Primary Completion Date',
            'Study Status',
            'Lead Sponsor',
            'Collaborators',
            'Study Locations',
            'Contacts',
            'References'
        ]

        self.df = {i: [] for i in key_words}
        for key_word in key_words:
            all_trial_data = self.get_all_trials(key_word)
            self.df[key_word] = self.extract_relevant_info(key_word, all_trial_data)

    def get_all_trials(self, key_word):
        all_trials = []
        min_rank = 1
        max_rank = 100  # ClinicalTrials.gov limits to 100 results per request

        for i in range(20):
            req = requests.get(self.CLINICAL_ENDPOINT(key_word, min_rank, max_rank))
            if req.status_code == 200:
                req = req.json()
                studies = req.get('FullStudiesResponse', {}).get('FullStudies', [])
                if not studies:
                    break
                all_trials.extend(studies)
                min_rank += 100
                max_rank += 100
            else:
                print(f"Failed to fetch trials for {key_word}. Status code: {req.status_code}")
                break

        return all_trials

    def extract_relevant_info(self, key_word, all_trial_data):
        extracted_data = []

        for study in all_trial_data:
            try:
                study_info = study.get('Study', {})
                protocol_section = study_info.get('ProtocolSection', {})
                contacts_location_section = study_info.get('ContactsLocationsModule', {})

                ncid = protocol_section.get('IdentificationModule', {}).get('NCTId', 'N/A')
                title = protocol_section.get('IdentificationModule', {}).get('OfficialTitle', 'N/A')
                recruitment_status = protocol_section.get('StatusModule', {}).get('OverallStatus', 'N/A')
                conditions = protocol_section.get('ConditionsModule', {}).get('ConditionList', {}).get('Condition', [])
                interventions = [intervention.get('InterventionName', 'N/A') for intervention in protocol_section.get('ArmsInterventionsModule', {}).get('InterventionList', {}).get('Intervention', [])]
                study_type = protocol_section.get('DesignModule', {}).get('StudyType', 'N/A')
                brief_summary = protocol_section.get('DescriptionModule', {}).get('BriefSummary', 'N/A')
                intervention_description = [intervention.get('InterventionDescription', 'N/A') for intervention in protocol_section.get('ArmsInterventionsModule', {}).get('InterventionList', {}).get('Intervention', [])]

                eligibility_criteria = protocol_section.get('EligibilityModule', {}).get('EligibilityCriteria', 'N/A').strip()

                primary_outcomes = protocol_section.get('OutcomesModule', {}).get('PrimaryOutcomeList', {}).get('PrimaryOutcome', [])
                secondary_outcomes = protocol_section.get('OutcomesModule', {}).get('SecondaryOutcomeList', {}).get('SecondaryOutcome', [])

                contacts = []
                for contact in contacts_location_section.get('CentralContactList', {}).get('CentralContact', []):
                    contacts.append({
                        'Name': contact.get('CentralContactName', 'N/A'),
                        'Phone': contact.get('CentralContactPhone', 'N/A'),
                        'Email': contact.get('CentralContactEMail', 'N/A')
                    })

                primary_outcomes_list = []
                for outcome in primary_outcomes:
                    primary_outcomes_list.append({
                        'Measure': outcome.get('PrimaryOutcomeMeasure', 'N/A'),
                        'Description': outcome.get('PrimaryOutcomeDescription', 'N/A'),
                        'Time Frame': outcome.get('PrimaryOutcomeTimeFrame', 'N/A')
                    })

                secondary_outcomes_list = []
                for outcome in secondary_outcomes:
                    secondary_outcomes_list.append({
                        'Measure': outcome.get('SecondaryOutcomeMeasure', 'N/A'),
                        'Description': outcome.get('SecondaryOutcomeDescription', 'N/A'),
                        'Time Frame': outcome.get('SecondaryOutcomeTimeFrame', 'N/A')
                    })

                start_date = protocol_section.get('StatusModule', {}).get('StartDateStruct', {}).get('StartDate', 'N/A')
                primary_completion_date = protocol_section.get('StatusModule', {}).get('PrimaryCompletionDateStruct', {}).get('PrimaryCompletionDate', 'N/A')
                study_status = protocol_section.get('StatusModule', {}).get('OverallStatus', 'N/A')
                lead_sponsor = protocol_section.get('SponsorCollaboratorsModule', {}).get('LeadSponsor', {}).get('LeadSponsorName', 'N/A')

                collaborators = []
                collaborator_list = protocol_section.get('SponsorCollaboratorsModule', {}).get('CollaboratorList', {}).get('Collaborator', [])
                for collab in collaborator_list:
                    collaborators.append(collab.get('CollaboratorName', 'N/A'))

                study_locations = []
                for location in contacts_location_section.get('LocationList', {}).get('Location', []):
                    study_locations.append({
                        'Facility': location.get('LocationFacility', 'N/A'),
                        'Status': location.get('LocationStatus', 'N/A'),
                        'City': location.get('LocationCity', 'N/A'),
                        'Country': location.get('LocationCountry', 'N/A')
                    })

                references = []
                for reference in protocol_section.get('ReferencesModule', {}).get('ReferenceList', {}).get('Reference', []):
                    references.append({
                        'PMID': reference.get('ReferencePMID', 'N/A'),
                        'Type': reference.get('ReferenceType', 'N/A'),
                        'Citation': reference.get('ReferenceCitation', 'N/A')
                    })

                extracted_data.append({
                    'NCTId': ncid,
                    'Title': title,
                    'Recruitment Status': recruitment_status,
                    'Conditions': conditions,
                    'Interventions': interventions,
                    'Study Type': study_type,
                    'Brief Summary': brief_summary,
                    'Intervention Description': intervention_description,
                    'Eligibility Criteria': eligibility_criteria,
                    'Primary Outcomes': primary_outcomes_list,
                    'Secondary Outcomes': secondary_outcomes_list,
                    'Start Date': start_date,
                    'Primary Completion Date': primary_completion_date,
                    'Study Status': study_status,
                    'Lead Sponsor': lead_sponsor,
                    'Collaborators': collaborators,
                    'Study Locations': study_locations,
                    'Contacts': contacts,
                    'References': references
                })
            except Exception as e:
                print(f"Error extracting data for study: {e}")

        return extracted_data


API_CTGOV = lambda x, min_rank=1, max_rank=100: f'https://ClinicalTrials.gov/api/query/full_studies?expr={x}&min_rnk={min_rank}&max_rnk={max_rank}&fmt=json'

# user_key_word_search = ['breast+cancer', 'colon+cancer', 'penile+cancer', 'cancer', 'typhoid', 'parkinson', 'alzheimers', 'hepatitis+c', 'aids', 'hiv']
user_key_word_search = ['breast+cancer', 'colon+cancer']
cancer_df_obj = ClinicalTrialDataset(user_key_word_search, API_CTGOV)

# Combine all datasets into a single list
combined_dataset = []
for key_word, dataset in cancer_df_obj.df.items():
    combined_dataset.extend(dataset)

# Example usage:
print(f"Total studies retrieved: {len(combined_dataset)}")
```
